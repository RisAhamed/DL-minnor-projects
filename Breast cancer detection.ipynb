```python
# splitting datasets
x = cancer_data.drop('Target',axis =1)
y = cancer_data['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 1243,test_size = 0.2)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from  tensorflow import keras
tf.random.set_seed(4)
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Flatten(input_shape=(30,)))
model.add(keras.layers.Dense(32,activation = 'relu',input_dim = x_train.shape[1]))
model.add(keras.layers.Dense(8,activation = 'relu'))
model.add(keras.layers.Dense(1,activation = 'sigmoid'))
```

```python
model.compile(optimizer= 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs =10,batch_size = 30,validation_split = 0.2)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(x_test,y_test)
print(accuracy)
```

```python
print('loss--',loss)
print('accuracy--',accuracy)
```
